```python
#Open the data Up, iterate through

total_months = 0
total_profit = 0
Average_change =[]
change_list = []
month_list = []
#[86000, 98000, 30000...]
    

with open(budget_data) as file:
    variable = csv.reader(file)
    print(variable)
    
    header = next(variable) 
    print(header)

    for row in variable:
        print(row[1])
        print(row[0])
        total_months = total_months + 1
        total_profit = total_profit + int(row[1])
        Average_change.append(int(row[1]))
        month_list.append(row[0])
```

```python
print(total_months)
print(Average_change[2])
print(month_list)
    
```

```python
for i in range(1,len(Average_change)):
    #print(Average_change[0])
    print(Average_change[i] - Average_change[i-1])
    change_list.append(Average_change[i] - Average_change[i-1])
```

```python
print("total profit is", total_profit)
print("total months are" , len(month_list))


avg_change = sum(change_list)/len(change_list)
index_max = change_list.index(max(change_list))
index_min = change_list.index(min(change_list))
index_max_month = month_list[index_max]
index_min_month = month_list[index_min]
print("Average change is", avg_change)
print ("Max value element : ", max(change_list), (index_max_month))
print ("Min value element : ", min(change_list), (index_min_month))
```
